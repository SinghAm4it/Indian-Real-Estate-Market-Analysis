```python
# 9) Function to convert price strings to Cr (Crores)

def convert_price_to_cr(price):
    """
    Converts price strings like:
      ₹48 L  →  0.48
      ₹1.5 Cr → 1.5
      80 Lakh → 0.8
      2 Crores → 2
    Returns price in Cr (float)
    """
    if pd.isna(price):
        return np.nan
    
    # Convert to string and clean
    price = str(price).replace("₹", "").replace(",", "").strip().lower()
    price = price.replace("lakhs", "l").replace("lakh", "l").replace("crores", "cr").replace("crore", "cr")
    
    # Extract numeric part
    numbers = re.findall(r"[0-9]*\.?[0-9]+", price)
    if not numbers:
        return np.nan
    
    value = float(numbers[0])
    
    # Convert based on unit
    if "cr" in price:
        return value
    elif "l" in price:
        return value / 100
    
# Apply conversion
df["Price_in_Cr"] = df["Price"].apply(convert_price_to_cr)

# Convert dtype to float
df["Price_in_Cr"] = df["Price_in_Cr"].astype("float")
```

```python
# 10) Extract BHK Type from Property Title

PATTERN = re.compile(
    r"\b(?P<num>\d+(?:\.\d+)?(?:\+?)?)\s*(?P<unit>BHK|BH|BK|B|RH|RK|R)\b",
    flags=re.IGNORECASE
)

def extract_bhk_type(title: str):
    s = str(title)
    m = PATTERN.search(s)
    if not m:
        return None
    num  = m.group("num").replace(" ", "")
    unit = m.group("unit").upper()
    # normalize units
    unit_norm = {
        "R": "B",        
        "RK": "BK",   
        "RH":"BH"
    }.get(unit, unit)       # keep BHK/RK/R as-is
    return f"{num} {unit_norm}"

df["BHK_Type"] = df["Property Title"].astype(str).apply(extract_bhk_type)

df["BHK_Type"].unique()

```

```python
# 11) Extract Property Type from Property Title

def extract_property_type(title, location=None):
    # Convert to lowercase string
    title = str(title).lower()

    pattern = re.compile(
        r"^\s*\d+(?:\.\d+)?\+?\s*(?:BHK|RK|R|B|BH|BK|RH)\b[\s-]*",
        flags=re.IGNORECASE
    )

    # Substitute the matched part (and everything before it) with an empty string
    title = re.sub(pattern, "", title).strip()
    
    # Remove everything after any comma or connector word
    title = re.sub(r"(\b(for)\b).*", "", title)

    # Capitalize properly
    return title.title().strip() if title else None


# Apply to DataFrame
df["Property_Type"] = df.apply(
    lambda row: extract_property_type(row["Property Title"], row["Location"]),
    axis=1
)

df["Property_Type"].unique()
```

```python
# 11) Split Location into Locality and City

df["Locality"] = df["Location"].astype(str).apply(lambda x: x.rsplit(",", 1)[0].strip() if "," in x else x.strip())
df["City"] = df["Location"].astype(str).apply(lambda x: x.rsplit(",", 1)[1].strip() if "," in x else None)

df["City"].unique()
```
